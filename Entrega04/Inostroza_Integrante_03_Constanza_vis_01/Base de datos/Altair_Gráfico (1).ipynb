```python
import pandas as pd
import altair as alt

# Gráfico 1: Edad Promedio de la Población Chilena
# Cargar el archivo CSV con delimitador ';'
data1 = pd.read_csv('/content/Datos_entrega04_limpio.csv', delimiter=';')
data1.columns = ['EdadPromedio', 'Anio']
data1['Anio'] = data1['Anio'].astype(int)
data1['EdadPromedio'] = pd.to_numeric(data1['EdadPromedio'], errors='coerce')

# Crear el gráfico con color y tooltip
chart1 = alt.Chart(data1).mark_bar().encode(
    x=alt.X('Anio:O', title='Año'),
    y=alt.Y('EdadPromedio:Q', title='Edad Promedio de la Población Chilena'),
    color=alt.Color('EdadPromedio:Q', scale=alt.Scale(scheme='spectral')),  # Usar una paleta personalizada
    tooltip=['Anio', 'EdadPromedio']  # Agregar tooltip con 'Año' y 'EdadPromedio'
).properties(
    title='Evolución de la Edad Promedio de la Población Chilena a lo Largo de los Años',
    width=500,
    height=300
)

# Gráfico 2: Porcentaje de adultos mayores de 60 años
# Cargar el archivo CSV con delimitador ';'
data2 = pd.read_csv('/content/datos_grafico_02.csv', delimiter=';')
data2.columns = ['porcentaje', 'Anio']
data2['Anio'] = data2['Anio'].astype(int)
data2['porcentaje'] = pd.to_numeric(data2['porcentaje'], errors='coerce')

# Crear el gráfico
chart2 = alt.Chart(data2).mark_bar().encode(
    x=alt.X('Anio:O', title='Año'),
    y=alt.Y('porcentaje:Q', title='Porcentaje adultos mayores de 60 años'),
    color=alt.Color('porcentaje:Q', scale=alt.Scale(scheme='spectral')),  # Usar una paleta personalizada
    tooltip=['Anio', 'porcentaje']  # Agregar tooltip con 'Año' y 'EdadPromedio'
).properties(
    title='Evolución del porcentaje de adultos mayores de 60 años en Chile',
    width=500,
    height=300
)

# Combinar ambos gráficos de forma horizontal
combined_chart = alt.hconcat(chart1, chart2)

# Mostrar el gráfico combinado
combined_chart

```
