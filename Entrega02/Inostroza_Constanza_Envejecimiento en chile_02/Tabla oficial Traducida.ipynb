```python
# transformar delimitad de punto y coma en csv a coma

import csv

# Archivo de entrada y salida
archivo_entrada = '/content/sample_data/TABLA oficial 2.0.csv'
archivo_salida = 'archivo_final_coma.csv'

# Abrir el archivo de entrada (con punto y coma) y leerlo
with open(archivo_entrada, mode='r', newline='', encoding='utf-8') as csv_entrada:
    lector = csv.reader(csv_entrada, delimiter=';')

    # Abrir el archivo de salida (con coma) y escribirlo
    with open(archivo_salida, mode='w', newline='', encoding='utf-8') as csv_salida:
        escritor = csv.writer(csv_salida, delimiter=',')

        # Copiar cada fila del archivo de entrada al archivo de salida
        for fila in lector:
            escritor.writerow(fila)



```
