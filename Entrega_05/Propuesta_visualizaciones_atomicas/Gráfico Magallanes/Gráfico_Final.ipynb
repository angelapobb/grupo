```python
import altair as alt
import pandas as pd

# Datos de ejemplo
data = pd.DataFrame({
    'Región': ['Arica y Parinacota', 'Tarapacá', 'Antofagasta', 'Atacama', 'Coquimbo', 'Valparaíso',
               'Metropolitana', 'O’Higgins', 'Maule', 'Ñuble', 'Biobío', 'La Araucanía', 'Los Ríos',
               'Los Lagos', 'Aysén', 'Magallanes'],
    'Cambio': [9, 6.3, 8.8, 7, 6.9, 8.7, 7.2, 9.3, 9.8, 11.6, 12.1, 11.4, 10.2, 8.5, 9.5, 11.9]
})

# Crear el gráfico con un solo color personalizado
chart = alt.Chart(data).mark_bar(color='#F58518').encode(  # Color personalizado único
    x=alt.X('Región:N', axis=alt.Axis(labelAngle=-45, labelFontSize=14, titleFontSize=16), sort=data['Región'].tolist()),  # Ordenar las barras según el orden de las regiones
    y=alt.Y('Cambio:Q', axis=alt.Axis(title="60 y más (%) cambio", labelFontSize=14, titleFontSize=16)),
    tooltip=[  # Añadir tooltip
        alt.Tooltip('Región:N', title="Región"),
        alt.Tooltip('Cambio:Q', title="Variación (%)")
    ]
).properties(
    title=alt.TitleParams(
        text="Variación % de la población +60 años, por región",
        fontSize=18,
        anchor='middle'
    ),
    width=800,
    height=400
)

# Agregar una línea horizontal que marque el promedio nacional
mean_line = alt.Chart(pd.DataFrame({'y': [8.1]})).mark_rule(
    color='blue',
    strokeWidth=4
).encode(
    y=alt.Y('y:Q', title=""),
    tooltip=[alt.Tooltip('y:Q', title="Promedio nacional")]
)

# Combinar el gráfico con la línea
final_chart = chart + mean_line

final_chart
```
